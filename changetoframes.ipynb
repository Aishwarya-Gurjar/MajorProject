```python
import cv2 
from PIL import Image
import numpy as np
import math as m
import os

def numberofsameframe(a):
    count=0
    for x in a:
        if x !=0:
            count+=1
    return count
    
def creating_folder_inside_sign(file):
    path = './segmented1/'+file
    files = os.listdir(path)
    nooffolder=len(files)+1
    os.mkdir('./segmented1/'+file+"/"+str(nooffolder))
    return nooffolder
    
def list_for_frame_accept(frames):
    value = frames
    z=frames
    counter =2
    while z < 30:
        z= value*counter
        counter+=1
    value =z
    counter -=1
    a = list(range(1,value+1,1))
    if len(a)>30:
        b = len(a)/(len(a)-30)
        x=0
        while x<=len(a)-m.floor(b):
            x=x+b
            e=int(m.floor(x))-1
            a[e]=0
        
        return a,counter
    else:
        return a,counter
    
def creating_folder_for_sign(listofsign):
    if not os.path.exists('segmented1/'+listofsign):
        os.mkdir('segmented1/'+listofsign)
    
    

def FrameCapture(fileoffolders): 
    list_of_file = os.listdir("made//"+fileoffolders+"//")
    print(fileoffolders)
    creating_folder_for_sign(fileoffolders)
    for file in list_of_file:
        print(file)
        vidObj = cv2.VideoCapture("made//"+fileoffolders+"//"+file) 
        counting_frames = 0
        success = 1
        while success:
            success, image = vidObj.read() 
            counting_frames += 1
        counting_frames-=1
        print(counting_frames)
        count =0
        a,counter=list_for_frame_accept(counting_frames)
        print(a)
        success = 1
        vidObj = cv2.VideoCapture("made//"+fileoffolders+"//"+file) 
        countofframe=1
        counter1=0 
        nooffolder=creating_folder_inside_sign(fileoffolders)
        while success: 
            success, image = vidObj.read() 
            for i in range(numberofsameframe(a[counter1*counter:(counter1+1)*counter])):
                cv2.imwrite("./segmented1/"+fileoffolders+"/"+str(nooffolder)+"/%d.jpg" % countofframe, image)              
                size = 440, 440
                image = Image.open(".//segmented1//"+fileoffolders+"//"+str(nooffolder)+"//%d.jpg" % countofframe) 
                area = (180, 0, 1140, 720)
                image = image.crop(area)
                image= image.resize(size, Image.ANTIALIAS)
                image.save(".//segmented1//"+fileoffolders+"//"+str(nooffolder)+"//%d.jpg" % countofframe)
                countofframe += 1
                count += 1
            counter1+=1

path = "made"
files = os.listdir(path)
# print(files)
for file in files: 
    FrameCapture(file) 
```
